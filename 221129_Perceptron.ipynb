```python
import torch
import torch.nn as nn
```

```python
device = 'cuda' if torch.cuda.is_available() else 'cpu'
```

```python
X = torch.FloatTensor([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
]).to(device)
Y = torch.FloatTensor([
    [0],[1],[1],[1]
]).to(device)
```

```python
model = nn.Sequential(
    nn.Linear(2,1,bias=True),
    nn.Sigmoid()
).to(device)
```

```python
loss_func = torch.nn.BCELoss().to(device)
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)
```

```python
for step in range(1000001):
  optimizer.zero_grad()

  Y_hat = model(X)
  loss = loss_func(Y_hat,Y)
  loss.backward()
  optimizer.step()

  if step % 10000 == 0:
    print(step, loss.item())
```

```python
with torch.no_grad():
  model.eval()
  inputs = torch.FloatTensor([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
]).to(device)
outputs = model(inputs)

print(outputs)
print(outputs > 0.5)
```
